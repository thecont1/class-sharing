```python
cnt = data['Genre'].value_counts()

x_bar = cnt.index
y_bar = cnt
plt.bar(x_bar, y_bar)
```

```python
plt.bar(x_bar, y_bar)
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.bar(x_bar, y_bar)
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.bar(x_bar, y_bar)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.bar(x_bar, y_bar)
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.bar(x_bar, y_bar, width=0.90, color='orange')
plt.title('Games per Genre')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (15, 8))
plt.bar(x_bar, y_bar, width=0.90)
plt.title('Games per Genre')
plt.xlabel('Genre', fontsize = 12)
plt.ylabel('Count', fontsize = 15)
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.countplot(x = 'Genre',
              data = data,
              order = data['Genre'].value_counts().index,
              color = 'cornflowerblue')
plt.xticks(rotation = 90)
plt.show()
```

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales_data
```

```python
region_sales = sales_data.T.sum(axis='columns')
region_sales
```

```python
plt.pie(region_sales)
```

```python
plt.pie(region_sales, labels = region_sales.index)
plt.show()
```

```python
plt.pie(region_sales,
        labels = region_sales.index,
        startangle = 90,
        autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(region_sales,
        labels = region_sales.index,
        startangle = 90,
        explode = [0.1, 0.1, 0, 0],
        autopct = '%0.1f%%')
plt.show()
```
