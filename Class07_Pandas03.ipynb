```python
data = movies.merge(directors, how='left', left_on='director_id', right_on='id')
data
```

```python
data.drop(['director_id','id_y'],axis=1, inplace = True)
data
```

```python
def encode(data):
  if data == "Male":
    return 'M'
  elif data == "Female":
    return 'F'
  else:
    return None
```

```python
data['sex_code'] = data['gender'].apply(encode)
data
```

```python
data[['revenue', 'budget']]
```

```python
data[['revenue', 'budget']].apply(np.sum)
```

```python
def profit(data):
  return data['revenue'] - data['budget']
```

```python
data.apply(profit, axis = 1)
```

```python
data['profit'] = data[['revenue', 'budget']].apply(profit, axis = 1)
data['profit']
```

```python
# Let's say, we want to filter the risky movies whose budget was even higher 
# than the average revenue of the director from all their other movies.
```

```python
def func(x):
  x['risky'] = x['budget'] - x['revenue'].mean() >= 0
  return x
```

```python
the_risky_women = data.groupby("director_name").apply(func, include_groups=False)
the_risky_women[the_risky_women['risky']][the_risky_women['sex_code']=='F']
```

```python
the_risky_women = data.groupby("director_name", group_keys=True).apply(func, include_groups=False).reset_index(drop=True)
the_risky_women[the_risky_women['risky']==True][the_risky_women['sex_code']=='F'].drop(['id_x', 'popularity', 'budget', 'revenue', 'vote_count', 'month', 'day', 'gender', 'risky'], axis=1)
```
