```python
df = pd.DataFrame({'name':["Ram","Shyam","Mukesh","Suresh"],'age':[10,18,60,50]})
df
```

```python
def categorize(age):
    if age>50:
        return 'senior'
    elif 18 <= age <= 50:
        return 'adult'
    elif age < 18:
        return 'kid'


df['age'] = df['age'].apply(categorize)
df
```

```python
data = {
    'name': ['Jim', 'Clarke', 'Kent', 'Mark'],
    'username': ['itsjimhere', 'clark002', 'itskentment', 'markyoumustknow'],
    'userid': [20, 10, 86, 21]
}

df = pd.DataFrame(data)
df
```

```python
def check(name, username):
    return name.lower() in username

df[~(df.apply(lambda x:check(x['name'],x['username']), axis=1))]['userid']
```
